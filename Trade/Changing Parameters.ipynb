```python
def plot_scatter_with_pnl(macro_result, y_axis_column, title, x_label, y_label):

    macro_result['date'] = pd.to_datetime(macro_result['date'])

    plt.figure(figsize=(12, 6))

    colors = ['green' if pnl >= 0 else 'red' for pnl in macro_result[y_axis_column]]

    plt.scatter(macro_result['date'], macro_result[y_axis_column], c=colors, s=100, edgecolor='k', alpha=0.75)

    plt.axhline(0, color='grey', linestyle='--', linewidth=1)

    plt.title(title, fontsize=16)
    plt.xlabel(x_label, fontsize=14)
    plt.ylabel(y_label, fontsize=14)
    plt.grid(True)

    plt.tight_layout()
    plt.show()
```

# Scenario Analysis #1

```python
upperbound_values = [0.5, 1, 1.5]
lowerbound_values = [-0.5, -1, -1.5]

# Initialize a list to store results
results = []
```

```python
for UPPERBOUND in upperbound_values:
    for LOWERBOUND in lowerbound_values:
        # Copy the original DataFrame
        macro_trial = macro_df.copy()
        
        # Apply trading strategy
        macro_trial = macro_trading_periods(macro_trial, UPPERBOUND, LOWERBOUND, MIN_CONSECUTIVE_DAYS, DRAWDOWN_THRESH)
        macro_trial = process_trading_periods(macro_trial)
        macro_trial['long_etf'] = np.where(
            (macro_trial['Z-Score'] > UPPERBOUND) & (macro_trial['Z-Score'] > 0), 'FEZ',
            np.where(
                (macro_trial['Z-Score'] < LOWERBOUND) & (macro_trial['Z-Score'] < 0), 'EWJ',
                np.nan
            )
        )
        macro_trial = calculate_trade_pnl(macro_trial, LONG)
        macro_result = create_trade_summary_df(macro_trial)
        
        # Calculate total return for the parameter combination
        total_return = macro_result['return'].sum()
        
        # Calculate risk metrics
        risk_metrics = calculate_risk_metrics(macro_result)
        
        # Store the results
        results.append({
            'UPPERBOUND': UPPERBOUND,
            'LOWERBOUND': LOWERBOUND,
            'total_return': total_return,
            'risk_metrics': risk_metrics
        })

# Convert results to a DataFrame for analysis
results_df = pd.DataFrame(results)

print(results_df)
```

```python
results_df['TOTAL_RETURN'] = results_df['risk_metrics'].apply(lambda x: round(x['Total Return'] / 1e6, 2))

# Create a pivot table for better plotting
pivot_table = results_df.pivot_table(
    index=['UPPERBOUND'],
    columns=['LOWERBOUND'],
    values='TOTAL_RETURN'
)


# Plot the heatmap
plt.figure(figsize=(14, 8))
sns.heatmap(pivot_table, annot=True, fmt=".2f", cmap='Blues', linewidths=0.5, cbar_kws={'label': 'Total Return (in Millions)'})
# plt.title('Total Returns for Each Parameter Combination', fontsize=16)
plt.xlabel('LOWERBOUND Z-Score', fontsize=14)
plt.ylabel('UPPERBOUND Z-Score', fontsize=14)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()
```
